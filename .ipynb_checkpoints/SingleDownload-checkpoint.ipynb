```python
def get_firstyear(h):
    firstday = h.index[0]
    if firstday.month > 6:
        firstyear = firstday.year+1
        print ("first year starts too late: "+str(firstday))
    else:
        firstyear = firstday.year
    return firstyear
```

retrieve last year for evaluation  
we need only values from the first halfyear but that should be complete (lastdate >= 30.06.)

```python
def get_lastyear(h):
    lastday = h.index[-1]
    if lastday.month < 6 and lastday.day < 30:
        lastyear = lastday.year-1
        print("last year ends too early: "+str(lastday))
    else:
        lastyear = lastday.year
    return lastyear
```

```python
results = pd.DataFrame(columns={'Symbol','Years','SellAVG','SellSTD','BuyAVG','BuySTD','DDAVG','DDMax','DDSTD','UpAVG','UpMax','UpSTD'})
results = results[['Symbol','Years','SellAVG','SellSTD','BuyAVG','BuySTD','DDAVG','DDMax','DDSTD','UpAVG','UpMax','UpSTD']]
```

```python
for i in range(selldates.index.size):
    #the solution for standard deviation from Timestamps:
    # now calculate the necessary values for the results
    symbol = selldates.iloc[i].name
    years = selldates.iloc[i].count()
    selldate_ints = selldates.iloc[i].astype(int)
    sell_avg = pd.to_datetime(selldate_ints.mean())
    sell_stddev= (pd.to_datetime(selldate_ints.std())-pd.to_datetime(0)).days
    
    buydate_ints = buydates.iloc[i].astype(int)
    buy_avg = pd.to_datetime(buydate_ints.mean())
    buy_stddev = (pd.to_datetime(buydate_ints.std())-pd.to_datetime(0)).days
    
    ddavg = drawdowns.iloc[i].mean()
    ddmax = drawdowns.iloc[i].max()
    ddstd = drawdowns.iloc[i].std()
    
    upavg = uplifts.iloc[i].mean()
    upmax = uplifts.iloc[i].max()
    upstd = uplifts.iloc[i].std()
    
    # go on with buyavg, buystd and so on
    results = results.append({'Symbol':symbol, 'Years': years, 'SellAVG': sell_avg, 'SellSTD':sell_stddev,
                             'BuyAVG': buy_avg, 'BuySTD': buy_stddev,
                             'DDAVG':ddavg, 'DDMax':ddmax, 'DDSTD':ddstd,
                             'UpAVG':upavg, 'UpMax':upmax, 'UpSTD':upstd},ignore_index=True)
```
